# Altomação de Tarefas

## Problemas a ser resolvido

- Buscar altomaticamente os dados das ações no site do Yahoo!
- Analizar os dados das ações de acordo com a solicitação do gestor
- Enviar um e-mail com todas as análises de forma altomática 

## Passo 1 - Buscar altomaticamente os dados das ações no site do Yahoo!

```python
import yfinance

codigo = str(input("Digite o codigo da ação desejada: "))
data = str(input("Qual a data dos dados: "))

dados = yfinance.Ticker(codigo).history(data)
dados

fechamento = dados.Close
print(fechamento)
print(fechamento.plot())
```

## Passo 2 - Analizar os dados das ações de acordo com a solicitação do gestor

- Analíse dos últimos 6 meses
- Cotação Máximo
- Cotação Mínimo
- Cotação Atual

```python
# cotação máxima
maxima = fechamento.max()

# cotação mínima
minima = fechamento.min()

# cotação atual
atual = fechamento.iloc[-1]
```

## Enviar um e-mail com todas as análises de forma altomática

- Abrir uma nova aba (Ctrl + T)
- Escrever o endereço de e-mail e apertar ENTER
- Clicar no botão escrever
- Preencher o destinatario, TAB
- Preencher o assunto, TAB
- Preencher o corpo do e-mail
- Clicar no botão ENVIAR


```python
import pyautogui
import pyperclip
import time

# Configurando uma pausa entre os passos
pyautogui.PAUSE = 2

# Abrir uma nova aba (Ctrl + T)
pyautogui.hotkey("ctrl", "t")

# Escrever o endereço de e-mail e apertar ENTER
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# Clicar no botão escrever
#pyautogui.position() <- para descobrir a posição do mause
time.sleep(5)
pyautogui.click(x=90, y=189)

# Preencher o destinatario, TAB
pyperclip.copy("email@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Preencher o assunto, TAB
pyperclip.copy("Análises Diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")


# Preencher o corpo do e-mail
mensagem = f"""
Prezado Gestor,

Segui as análises dos últimos 6 meses da ação {codigo} conforme solicitado: 

Cotação máxima: R$ {round(maxima, 2)}
Cotação mínima: R$ {round(minima, 2)}
Cotação atual: R$ {round(atual, 2)}

Qualquer dúvida, estou á disposição!

Atte.
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Clicar no botão ENVIAR
pyautogui.click(x=843, y=693)
```
